```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PROJECT/Dataset/train.csv")
```

```python
data.head()
```

```python
0data['Order Date'] = pd.to_datetime(data['Order Date'], format="%d/%m/%Y")
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format="%d/%m/%Y")
data["Processing Time"] = (data["Ship Date"] - data["Order Date"]).dt.days
median_processing_time = data.groupby("Sub-Category")["Processing Time"].median().sort_values()
print("Median processing time for each sub-category")
print(median_processing_time)
```

```python
plt.figure(figsize=(12, 6))
median_processing_time.plot(kind="bar", color='yellow')
plt.xlabel("Sub-Category")
plt.ylabel("Median Processing Time (Days)")
plt.title("Median Processing Time for Each Sub-Category")
plt.show;
```
